### Kerenl PCA Function 

```python
def Stepwise_KPCA(X,gamma,n_components):
    d_matrix = euclidean_distances(X)
    d_matrix_2 = np.power(d_matrix,2)
    K = np.exp(-1*gamma*d_matrix_2)

    one_matrix = np.ones((X.shape[0], X.shape[0])) / X.shape[0]
    one_K = one_matrix.dot(K)
    K_one = K.dot(one_matrix)
    one_K_one = one_matrix.dot(K).dot(one_matrix)

    Centered_K = K - one_K - K_one + one_K_one

    eigen_vals, eigen_vecs = np.linalg.eigh(Centered_K)

    sorted_indices = np.argsort(eigen_vals)[::-1]
    sorted_eigen_vals = eigen_vals[sorted_indices] 
    sorted_eigen_vecs = eigen_vecs[:, sorted_indices]

    w = np.hstack([sorted_eigen_vecs[:, i].reshape(-1, 1) for i in range(n_components)])

    return w
```

### Training data onto KPCA Function with One Component

```python
moon_1d = Stepwise_KPCA(X_moon,gamma=15,n_components=1)
moon_1d.shape
```
