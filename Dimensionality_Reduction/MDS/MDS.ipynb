# Multi-Dimension Scaling

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import manhattan_distances
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.manifold import MDS
```

### Reading Data through Pandas

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/''machine-learning-databases/wine/wine.data',header=None)
df_wine.columns = [
    'Class label', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
    'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
    'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'
]
```

```python
df_wine.head()
```

### Splitting Dataset into X(features) and y(labels)

```python
X,y = df_wine.iloc[:,1:],df_wine.iloc[:,0]
```

## MDS through sklearn

### Calculating Manhatten Distance Matrix

```python
d_matrix = manhattan_distances(X)
d_matrix
```

### Training MDS on Manhatten Distance Matrix using Sklearn

```python
mds = MDS(n_components=2,normalized_stress=False,metric=True,dissimilarity="precomputed",random_state=0,eps=1e-9)
mds_2d = mds.fit_transform(d_matrix)
```

### Plotting Transformed Datapoints

```python
colors = ["r","g","b"]
markers = ["o","^","s"]
for l,c,m in zip(np.unique(y),colors,markers):
    plt.scatter(mds_2d[y==l,0],mds_2d[y==l,1],c=c,label=f"Class {l}",marker=m)
plt.xlabel("MDS 1")
plt.ylabel("MDS 2")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

### Calculating Euclidean Distance Matrix

```python
d_matrix = euclidean_distances(X)
d_matrix
```

### Training MDS on Euclidean Distance Matrix using Sklearn

```python
mds_2d = mds.fit_transform(d_matrix)
```

### Plotting Transformed Datapoints

```python
colors = ["r","g","b"]
markers = ["o","^","s"]
for l,c,m in zip(np.unique(y),colors,markers):
    plt.scatter(mds_2d[y==l,0],mds_2d[y==l,1],c=c,label=f"Class {l}",marker=m)
plt.xlabel("MDS 1")
plt.ylabel("MDS 2")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

### Computing Transformed Datapoints Euclidean Matrix 

```python
d_embedded = euclidean_distances(mds_2d)
d_embedded
```

### Plotting Shepard Diagram 

```python
plt.figure(figsize=(8, 6))
plt.scatter(d_matrix, d_embedded, alpha=0.7, label='Point pairs')
plt.xlabel("Original Dissimilarities")
plt.ylabel("Reduced Space Distances")
plt.title("Shepard Diagram")
plt.legend()
plt.grid(True)
plt.show()
```

## Classical MDS

### Computing D^2(euclidean matrix ** 2)

```python
G = d_matrix**2
G
```

### Calculating H

```python
r = np.ones((1,X.shape[0]))
I = np.identity(X.shape[0])
H = I - ((1/X.shape[0])*(r.T.dot(r)))
H
```

### Calculating Gram Matrix

```python
G_center = -0.5*(H.dot(G).dot(H))
G_center
```

### Calculating Eigen Values and Eigen Vectors

```python
eigen_values,eigen_vector = np.linalg.eig(G_center)
```

### Sorting Eigen Values and Eigen Vectors

```python
indices = np.argsort(eigen_values)[::-1]
eigen_values = eigen_values[indices]
eigen_vector = eigen_vector[:,indices]
ev = eigen_vector[:,:2]
eg = eigen_values[:2]
```

### Calculating Transformed Datapoints

```python
L_k = np.diag(np.sqrt(eg))
Y = ev.dot(L_k)
Y_real = Y.real ## As y results in imaginary parts so only considering real part only
```

### Plotting Transformed Datapoints

```python
colors = ["r","g","b"]
markers = ["o","^","s"]
for l,c,m in zip(np.unique(y),colors,markers):
    plt.scatter(Y_real[y==l,0],Y_real[y==l,1],c=c,label=f"Class {l}",marker=m)
plt.xlabel("MDS 1")
plt.ylabel("MDS 2")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

### Calulating Euclidean Matrix for Transformed Datapoints

```python
dist_matrix_2 = euclidean_distances(Y_real)
```

### Plotting Shepard Diagram

```python
plt.figure(figsize=(8, 6))
plt.scatter(d_matrix, dist_matrix_2, alpha=0.7, label='Point pairs')
plt.xlabel("Original Dissimilarities")
plt.ylabel("Reduced Space Distances")
plt.title("Shepard Diagram")
plt.legend()
plt.grid(True)
plt.show()
```
