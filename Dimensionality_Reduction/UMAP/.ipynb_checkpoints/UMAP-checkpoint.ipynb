### Importing Libraries

```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import umap.umap_ as umap
from mpl_toolkits.mplot3d import Axes3D
```

### Loading Data

```python
digits = load_digits()
```

### Displaying Dataset Images

```python
fig,ax = plt.subplots(1,4)
for i in range(4):
    ax[i].imshow(digits.images[i],cmap = 'Greys')
```

### Splitting Dataset into X and y

```python
X_digits = digits.data
y_digits = digits.target
```

### Training UMAP Algorithm with Two Components

```python
Umap = umap.UMAP(n_components=2,min_dist=0.1,metric="euclidean",n_neighbors = 50)
umap_2d = Umap.fit_transform(X_digits)
```

### Plotting Transformed Datapoints

```python
def plot_projection(x,colors):
    plt.figure(figsize=(10, 8))
    for i in range(0,10):
        plt.scatter(x[colors==i,0],x[colors==i,1])
    for i in range(0,10):
        xtext,ytext = np.median(x[colors==i,:],axis=0)
        txt = plt.text(xtext,ytext,str(i),fontsize=24)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5,foreground="w"),PathEffects.Normal()])
```

```python
plot_projection(umap_2d,y_digits)
```

### Training UMAP Algorithm with Three Components

```python
Umap = umap.UMAP(n_components=3,min_dist=0.1,metric="euclidean",n_neighbors = 50)
umap_3d = Umap.fit_transform(X_digits)
```

### Plotting Transformed Datapoints

```python
def plot_projection_3d(x,colors):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    for i in range(0,10):
        ax.scatter(x[colors==i,0],x[colors==i,1],x[colors==i,2])
    for i in range(0,10):
        xtext,ytext,ztext = np.median(x[colors==i,:],axis=0)
        txt = ax.text(xtext,ytext,ztext,str(i),fontsize=24)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5,foreground="w"),PathEffects.Normal()])
```

```python
plot_projection_3d(umap_3d,y_digits)
plt.show()
```
