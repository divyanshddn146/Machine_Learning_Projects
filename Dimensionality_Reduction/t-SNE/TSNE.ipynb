### Importing Libraries

```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import matplotlib.patheffects as PathEffects
```

### Loading Dataset

```python
digits = load_digits()
```

### Displaying Dataset Images

```python
fig,ax = plt.subplots(1,4)
for i in range(4):
    ax[i].imshow(digits.images[i],cmap = 'Greys')
```

### Splitting Dataset into X and y

```python
X_digits = digits.data
y_digits = digits.target
```

### Training T-SNE Algorithm with Two Components

```python
tsne = TSNE(n_components=2,init='pca',random_state=0)
X_digit_tsne = tsne.fit_transform(X_digits)
```

### Plotting Transformed Datapoints

```python
def plot_projection(x,colors):
    plt.figure(figsize=(10, 8))
    for i in range(0,10):
        plt.scatter(x[colors==i,0],x[colors==i,1])
    for i in range(0,10):
        xtext,ytext = np.median(x[colors==i,:],axis=0)
        txt = plt.text(xtext,ytext,str(i),fontsize=24)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5,foreground="w"),PathEffects.Normal()])
```

```python
plot_projection(X_digit_tsne,y_digits)
```

### Training T-SNE Algorithm with Three Components

```python
tsne_3d = TSNE(n_components=3,init='pca',random_state=0,perplexity=25)
X_digit_tsne_3d = tsne_3d.fit_transform(X_digits)
```

### Plotting Transformed Datapoints

```python
def plot_projection_3d(x,colors):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    for i in range(0,10):
        ax.scatter(x[colors==i,0],x[colors==i,1],x[colors==i,2])
    for i in range(0,10):
        xtext,ytext,ztext = np.median(x[colors==i,:],axis=0)
        txt = ax.text(xtext,ytext,ztext,str(i),fontsize=24)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5,foreground="w"),PathEffects.Normal()])
```

```python
plot_projection_3d(X_digit_tsne_3d,y_digits)
plt.show()
```
