### Dividing Data into Training and Test Dataset

```python
indices = list(df_classif.index)   ### Retreiving Indices
random.seed(0)
random.shuffle(indices)            ### Randomizing indices

split_idx = int(0.7 * len(indices))   ### Splitting indices in ratio 70-30

train_idx = indices[:split_idx]       ### Getting Train indexes
test_idx = indices[split_idx:]        ### Getting Test indexes

train_classif = df_classif.loc[train_idx].reset_index(drop=True)  ### Training Data for Classification
test_classif  = df_classif.loc[test_idx].reset_index(drop=True)   ### Testing Data for Classification
```

```python
indices = list(df_regr.index)        ### Retreiving Indices
random.seed(0)
random.shuffle(indices)              ### Randomizing indices

split_idx = int(0.7 * len(indices))   ### Splitting indices in ratio 70-30

train_idx = indices[:split_idx]       ### Getting Train indexes
test_idx = indices[split_idx:]        ### Getting Test indexes

train_regr = df_regr.loc[train_idx].reset_index(drop=True)   ### Training Data for Regression
test_regr  = df_regr.loc[test_idx].reset_index(drop=True)    ### Testing Data for Regression
```

```python
# Splitting Data into X and y
X_train,y_train = train_classif.iloc[:,:-1],train_classif.iloc[:,-1] 
X_test,y_test = test_classif.iloc[:,:-1],test_classif.iloc[:,-1]
```

## Main KNN Algorithm

```python
# KNN Algorithm
class KNN_Algorithm:
    
    def __init__(self,k,metric,weights,mode):
        self.k = k                                            ### Number of nearest neighbour 
        self.metric = metric                                  ### what type of distance metric to use
        self.weights = weights                                ### what type of weights to use
        self.mode = mode                                      ### if you wantto do regression or classification
    
    def fit(self,X,y):                                        ### In KNN no training of data is required 
        self.X = np.array(X,copy=True)                        ### In Knn we only converts X and y into numpy arrays
        self.y = np.array(y,copy=True)
    
    def euclidean_dist(self,x1,x2):                          ### Calculating Euclidean Distance 
        return np.sqrt(np.sum((x1-x2)**2,axis=2))
        
    def manhattan_dist(self,x1,x2):                         ### Calculating Manhattan Distance
        return np.sum(np.abs(x1-x2),axis=2)
        
    def minkowski_dist(self, x1, x2, p=3):                  ### Calculating Minowski Distance
        return np.sum(np.abs(x1 - x2)**p, axis=1)**(1/p)
    
    def predict(self,X):
        
        X = np.array(X, copy=True)
        
        if self.metric=="euclidean":
            dists = self.euclidean_dist(X[:,np.newaxis,:],self.X[np.newaxis,:,:])     
        
        elif self.metric=="manhattan":
            dists = self.manhattan_dist(X[:,np.newaxis,:],self.X[np.newaxis,:,:])
        
        elif self.metric == "minkowski":
            dists = self.minkowski_dist(X[:,np.newaxis,:],self.X[np.newaxis,:,:])
            
        indices = np.argsort(dists, axis=1)[:,:self.k]
        k_labels = self.y[indices]                                                    ### Array containing Labels with columns k
        k_dists = np.sort(dists,axis=1)[:,:self.k]                                    ### Array containing Distances with columns k
         
        if self.weights == "uniform":                                                
            
            if self.mode == "classification":
                predictions = mode(k_labels, axis=1)[0].flatten()
            
            elif self.mode == "regression":
                predictions = np.sum(k_labels, axis=1)
                predictions = predictions/self.k
        
        elif self.weights == "distance":
            weights = 1 / (k_dists + 1e-5)
            
            if self.mode == "classification":
                weighted_votes = np.zeros_like(k_labels, dtype=float)
                for i in range(k_labels.shape[0]):
                    for j in range(self.k):
                        weighted_votes[i,k_labels[i,j]] += weights[i,j]
                predictions = np.argmax(weighted_votes, axis=1)
            
            elif self.mode == "regression":
                weighted_votes = np.zeros((k_labels.shape[0],1), dtype=float)
                for i in range(k_labels.shape[0]):
                    for j in range(self.k):
                         weighted_votes[i] += weights[i,j]*k_labels[i,j]
                        
                weighted_votes /= np.sum(weights,axis=1,keepdims=True)
                predictions = weighted_votes
                
                
        return predictions
```

## Decision Boundary Function

```python
def plot_decision_regions(X_train,y_train,X_test, y_test, model, resolution=0.02):
    
    # Generate the mesh grid
    x_min, x_max = X_train.iloc[:, 0].min() - 1, X_train.iloc[:, 0].max() + 1
    y_min, y_max = X_train.iloc[:, 1].min() - 1, X_train.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    
    # Reshape the meshgrid points into a 2D array
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    
    # Use the knn_algorithm to classify the grid points
    Z = model.predict(grid_points)
    
    # Reshape the result into the shape of the grid
    Z = np.array(Z).reshape(xx.shape)
    
    # Plot the decision boundary and scatter points
    fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(12,6))
    ax[0].contourf(xx,yy,Z,alpha=0.3,cmap=plt.cm.Paired)
    ax[0].scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolor="k")
    ax[1].contourf(xx,yy,Z,alpha=0.3,cmap=plt.cm.Paired)
    ax[1].scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=y_test, cmap=plt.cm.coolwarm, edgecolor="k")
    ax[0].set_xlabel("Feature_1")
    ax[0].set_ylabel("Feature_2")
    ax[0].set_title("Classification Scatter Plot for Training Set")
    ax[1].set_xlabel("Feature_1")
    ax[1].set_ylabel("Feature_2")
    ax[1].set_title("Classification Scatter Plot for Test Set")
    plt.show()
```

## Plotting Decision Boundary 

```python
### KNN Algorithm with 3 nearest neighbour
knn = KNN_Algorithm(k=3, metric="euclidean", weights="uniform",mode="classification")
knn.fit(X_train, y_train)
plot_decision_regions(X_train,y_train,X_test,y_test,knn)
```

```python
### KNN Algorithm with 3 nearest neighbour with weights
knn2 = KNN_Algorithm(k=3, metric="euclidean", weights="distance",mode="classification")
knn2.fit(X_train, y_train)
plot_decision_regions(X_train,y_train,X_test,y_test,knn2)
```

```python
### KNN Algorithm with 10 nearest neighbour
knn = KNN_Algorithm(k=10, metric="euclidean", weights="uniform",mode="classification")
knn.fit(X_train, y_train)
plot_decision_regions(X_train,y_train,X_test,y_test,knn)
```

```python
### KNN Algorithm with 3 nearest neighbour with weights
knn2 = KNN_Algorithm(k=10, metric="euclidean", weights="distance",mode="classification")
knn2.fit(X_train, y_train)
plot_decision_regions(X_train,y_train,X_test,y_test,knn2)
```
