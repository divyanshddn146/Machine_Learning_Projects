For the analysis of customer behavior based on their shopping times—**Early Birds**, **Night Shoppers**, and **Weekend Shoppers**—we can explore various types of visualizations and clustering techniques to better understand customer patterns. This includes graphical representations, statistical analyses, and segmentation models. Below is a detailed guide on what we can do in this context:

### **Key Features for Analysis**
- **hour**: The hour of the day when the customer places the order.
- **weekday**: The day of the week when the order is placed (could be used to identify trends across different days).
- **is_weekend**: Whether the order is placed during the weekend (true/false).
- **time_slot**: A predefined time period (morning, afternoon, night, etc.).

---

### **1. Visualizations for Early Birds, Night Shoppers, and Weekend Shoppers**

#### **A. Distribution of Orders by Hour**
- **Purpose**: Identify peak shopping times and categorize customers based on when they prefer to shop.
- **Graph Type**: **Bar Plot** or **Line Plot**
  - **X-axis**: `hour` (0-23, representing hours of the day)
  - **Y-axis**: Count of orders (or percentage of orders placed in each hour)
  - **Analysis**: This will show when most customers are placing orders and help you segment them into "Early Birds" (e.g., orders placed between 6 AM - 12 PM) and "Night Shoppers" (e.g., orders placed between 6 PM - 12 AM).

#### **B. Distribution of Orders by Weekday and Time of Day**
- **Purpose**: Identify if customers have a particular day or time preference.
- **Graph Type**: **Heatmap** or **Stacked Bar Plot**
  - **X-axis**: `weekday` (0 to 6 for Monday to Sunday)
  - **Y-axis**: `hour` (0-23)
  - **Color/Intensity**: Order count (or percentage)
  - **Analysis**: This can help you understand if certain days (weekends or weekdays) show a higher tendency for "Night Shoppers" or "Early Birds". You can see if certain hours are clustered around weekends or weekdays.

#### **C. Distribution of Weekend vs. Weekday Orders**
- **Purpose**: Identify the shopping behavior of customers on weekends vs weekdays.
- **Graph Type**: **Pie Chart** or **Bar Plot**
  - **X-axis**: `is_weekend` (True/False)
  - **Y-axis**: Count of orders (or customer counts)
  - **Analysis**: This will help segment customers who are more likely to shop on weekends. "Weekend Shoppers" will be a group with a significantly higher number of orders on weekends.

#### **D. Time Slot Preferences**
- **Purpose**: Identify whether customers tend to shop more in the morning, afternoon, or night.
- **Graph Type**: **Pie Chart** or **Bar Plot**
  - **X-axis**: `time_slot` (e.g., Morning, Afternoon, Night)
  - **Y-axis**: Count of orders or customer counts
  - **Analysis**: Segment customers based on their preferred time slots (morning or night), which will help further categorize them as "Early Birds" or "Night Shoppers".

#### **E. Heatmap of Time Slot by Weekday**
- **Purpose**: Understand the relationship between time slots and weekdays.
- **Graph Type**: **Heatmap**
  - **X-axis**: `weekday`
  - **Y-axis**: `time_slot`
  - **Color/Intensity**: Number of orders
  - **Analysis**: This helps visualize whether certain time slots (morning, afternoon, evening, night) are more popular on specific weekdays, helping you understand customer behavior in a detailed manner.

---

### **2. Customer Segmentation using Clustering**

To further segment customers based on when they shop, clustering algorithms can be applied to group customers who exhibit similar behaviors in terms of shopping times. Here’s how you can approach clustering:

#### **A. K-Means Clustering**
- **Goal**: Segment customers into distinct groups based on their time-based behaviors.
- **Features to Use**: `hour`, `weekday`, `is_weekend`, `time_slot`
- **Steps**:
  1. **Data Preprocessing**: Scale or normalize time-related features (e.g., `hour` and `weekday`) for clustering. One-hot encode `time_slot` and `is_weekend` if they are categorical.
  2. **Clustering**: Apply K-Means with an optimal number of clusters. You can use the **Elbow Method** or **Silhouette Score** to determine the right `k` value.
  3. **Analysis**: Once the clusters are formed, each cluster will represent a unique shopping behavior (e.g., "Early Birds", "Night Shoppers", and others).
  4. **Visualization**: Use 2D or 3D scatter plots to visualize the clustering results based on `hour`, `weekday`, and `time_slot`.

#### **B. Hierarchical Clustering**
- **Goal**: Discover clusters of customers based on their order timings.
- **Steps**:
  1. Use hierarchical clustering (Agglomerative Clustering) to build a dendrogram, which helps visualize how customers are grouped at different levels of similarity.
  2. Cut the dendrogram at a certain threshold to get the desired number of clusters (e.g., three clusters: Early Birds, Night Shoppers, and Mixed).
  3. **Visualization**: Dendrogram plot to visualize the customer grouping.

#### **C. DBSCAN (Density-Based Spatial Clustering)**
- **Goal**: Identify clusters of customers with similar time patterns, especially when you have a lot of noise (outliers).
- **Steps**:
  1. DBSCAN will group customers who have dense clusters based on shopping times.
  2. It works well when you have irregular clusters or noise (customers who don't follow any specific pattern).
  3. **Visualization**: Use scatter plots to show the clustered groups. Outliers will be marked as noise points.

### **3. Segmentation Profiles and Insights**

Once you have your clusters (e.g., Early Birds, Night Shoppers, Weekend Shoppers), you can analyze the following:

1. **Customer Demographics (if available)**: Investigate if there are any demographic factors that influence the shopping time preference. Are "Early Birds" more likely to be of a certain age group or location?
   
2. **Spending Patterns**:
   - Analyze if "Early Birds" tend to spend more than "Night Shoppers" or if certain time slots have higher average order amounts.
   - Use metrics like average `total_items`, `discount%`, or `order` value to compare spending behavior across time-based clusters.

3. **Marketing Opportunities**:
   - **Early Birds**: They might respond better to promotions sent early in the day. They could be more likely to engage with offers related to morning-specific products (e.g., breakfast items, coffee).
   - **Night Shoppers**: Target them with evening-specific offers. They may respond to discounts on items purchased later in the day.
   - **Weekend Shoppers**: Special promotions for weekends, including limited-time offers, could be effective.

4. **Retention Strategies**: 
   - For each segment, identify strategies to improve customer retention. For example, if "Night Shoppers" are more likely to place orders late, you might introduce loyalty rewards or promotions targeted at this behavior.

5. **Customer Lifetime Value (CLV)**:
   - Calculate CLV for each cluster to determine which segment is the most valuable long-term.

### **4. Advanced Analysis**

#### **A. Predictive Analytics**
- Once segments are identified, you can build predictive models to forecast customer behavior. For example:
  - **Churn Prediction**: Use the time-based segments to predict which customers are likely to stop buying based on their shopping times.
  - **Cross-Sell/Up-Sell Opportunities**: Analyze if certain time segments tend to buy complementary products at different times (e.g., "Night Shoppers" may purchase more snacks or drinks).

#### **B. Time Series Analysis**
- If you have historical order data, you can perform **time series analysis** to predict future shopping patterns or identify seasonal trends (e.g., increased shopping on weekends, holidays, or specific times of the day).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv("ulabox_orders.csv")
```

```python
df
```

```python
def assign_time_slot(hour):
    if 0 <= hour < 6:
        return "late night"
    elif 6 <= hour < 12:
        return "morning"
    elif 12 <= hour < 18:
        return "afternoon"
    elif 18 <= hour < 24:
        return "evening"
    else:
        return "unknown"  # In case of invalid values

df['time_slot'] = df['hour'].apply(assign_time_slot)
```

```python
df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)
```

##  Distribution of Orders by Hour

```python
hourly_orders = df.groupby('hour').size().reset_index(name='order_count')
hourly_orders.head()
```

```python
def map_to_time_category(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Late Night'

hourly_orders['time_category'] = hourly_orders['hour'].apply(map_to_time_category)
```

```python
time_category_orders = hourly_orders.groupby('time_category')['order_count'].sum().reset_index()

time_category_orders['time_category'] = pd.Categorical(
    time_category_orders['time_category'],
    categories=['Morning', 'Afternoon', 'Evening', 'Late Night'], 
    ordered=True
)
```

```python
sns.set_theme(style="darkgrid")

x = hourly_orders.iloc[:, 0]  # Hour
y = hourly_orders.iloc[:, 1]  # Order Count

marker_indices_1 = [4] 
marker_indices_2 = [22]

sns.lineplot(x=x, y=y, color='deepskyblue', linewidth=3,zorder=1)

plt.scatter(x[marker_indices_1], y[marker_indices_1], color='red', s=100, label="Marker at lowest Orders",zorder=2)
plt.scatter(x[marker_indices_2], y[marker_indices_2], color='green', s=100, label="Marker at Highest Orders",zorder=3)

plt.xlabel("Hour")
plt.ylabel("Order Count")
plt.legend()
plt.figure(figsize=(10, 8))
plt.show()
```

## Segmenting customers into distinct groups based on their time-based behaviors

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

features = ['hour', 'weekday', 'is_weekend', 'time_slot']

data = df[features]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['hour', 'weekday']),  # Scale numerical features
        ('cat', OneHotEncoder(), ['time_slot'])         # One-hot encode categorical
    ],
    remainder='passthrough'  # Keep 'is_weekend' as is
)

X = preprocessor.fit_transform(data)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine the optimal number of clusters using the Elbow Method
sse = []  # Sum of Squared Errors
k_values = range(2, 10)  # Test for 2 to 10 clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42,n_init="auto")
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, sse, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('SSE')
plt.title('Elbow Method to Find Optimal k')
plt.show()
```

```python

# Choose the optimal k based on the Elbow Method or try silhouette analysis
optimal_k = 6  # Adjust based on the graph
kmeans = KMeans(n_clusters=optimal_k, random_state=42,n_init="auto")
clusters = kmeans.fit_predict(X)

# Add cluster labels to the original data
df['time_cluster'] = clusters
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df['hour'], y=df['weekday'], 
    hue=df['time_cluster'], palette='Set2', s=100
)
plt.title('Customer Clusters by Hour and Weekday')
plt.xlabel('Hour')
plt.ylabel('Weekday')
plt.legend(title='Cluster')
plt.show()
```

```python
# 0.7 and 1000
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=3000)
clusters = dbscan.fit_predict(X)

# Add cluster labels to the original data
df['time_cluster'] = clusters
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df['hour'], y=df['weekday'], 
    hue=df['time_cluster'], palette='Set2', s=100
)
plt.title('Customer Clusters by Hour and Weekday')
plt.xlabel('Hour')
plt.ylabel('Weekday')
plt.legend(title='Cluster')
plt.show()
```
